```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data =pd.read_csv('student_scores.csv')
data
```

```python
y=data['Scores'].values
x=data['Hours'].values
x,y
```

```python
plt.figure(figsize=(8,6))
plt.title("Data Distrubation")
plt.scatter(x,y)
plt.xlabel("hours")
plt.ylabel("score")
plt.show()
```

```python
l_rate=0.001
iteration=100

theta_0=0
theta_1=1

n=x.shape[0]

losses=[]
for i in range(iteration):
    h_x=theta_0+theta_1*x

    mse=(1/n)* np.sum((h_x-y)**2)
    losses.append(mse)
    print("iteration :",i+1)
    print("loss:",mse)
    d_theta_0=(2/n)* np.sum(h_x-y)
    d_theta_1=(2/n)* np.sum(x*(h_x-y))

    theta_0=theta_0-l_rate*d_theta_0
    theta_1=theta_1-l_rate*d_theta_1
```

```python
print("theta_0",theta_0)
print("theta_1",theta_1)
print("mse",mse)
```

```python
new_x=9
predict=theta_0+theta_1*new_x
print("Score",predict)
```

```python
plt.title("Loss values")
plt.plot(losses)
plt.ylabel("loss")
plt.xlabel("iteration")
```
